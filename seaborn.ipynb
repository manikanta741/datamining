```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
tip=sns.load_dataset('tips')
print(tip.head())
```

```python
import seaborn as sns
sns.displot(tip['total_bill'])
plt.show()
```

```python
sns.jointplot( x='total_bill',y='tip',data=tip)
plt.show()
```

```python
sns.scatterplot( x='total_bill',y='tip',data=tip)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=True)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tip)
plt.show()
```
