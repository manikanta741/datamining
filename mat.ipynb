```python
import numpy as np
import pandas as pd
```

```python
#create an array for x value
x=np.linspace(10,200,50)
print(x)
```

```python
#define linear equation
m=2
c=3
y=m*x+c
print(y)
```

```python
#plot the equation
import matplotlib.pyplot as plt
plt.plot(x,y,label='y=m*x+c')
```

```python
#plot linear equation
plt.plot(x,y,label='y=2*x+3')
plt.title('plot of linear equation')
plt.xlabel('x' )
plt.ylabel('y')
plt.xlim(10,100)#x-axis limit
plt.ylim(10,100)#y-axis limit
plt.grid(True)#grids in plot
plt.legend()#label equation
plt.savefig('linearr..png')
plt.show()
```

```python
##matplot fig
#create array of x values
x=np.linspace(0,10,100)
#calcualte y vlues based on x
y=np.sin(x)
#create anew fig with specific size
fig=plt.figure(figsize=(8,6))
#add an axes to fig
ax=fig.add_axes([0.9,0.9,0.9,0.5])
#plot data on axes
ax.plot(x,y,label='sin(x)',color='blue')
#set title
ax.set_title('simple plot of sin(x)')
plt.legend()
#set x,y axis label
ax.set_xlabel('x')
ax.set_ylabel('applitude')
plt.show()
```
