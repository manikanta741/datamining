```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
data = np.random.normal(0, 1, 100)

# Create a boxplot
plt.boxplot(data)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
data = np.random.normal(0, 1, 100)

# Create a boxplot
plt.boxplot(data,
            labels=['Data 1', 'Data 2'],

            vert=True,
            patch_artist=True,
            boxprops={'color': 'black', 'linewidth': 2},
            flierprops={'color': 'red', 'marker': 'o'},
            medianprops={'color': 'green', 'linewidth': 2},
            meanprops={'color': 'yellow', 'linewidth': 2},
            showmeans=True,
            showmedians=True,
            whis=1.5,
            notch=False,
            bootstrap=10000,
            grid=True,
            widths=0.5,
            positions=[1, 2])

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample correlation matrix
corr_matrix = np.array([[1, 0.7, 0.3], [0.7, 1, 0.5], [0.3, 0.5, 1]])

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)

# Show plot
plt.show()
```
