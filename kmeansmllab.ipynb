```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
data,labels=make_blobs(n_samples=300,n_features=2,centers=5,random_state=42)
df=pd.DataFrame(data)
df.head()
```

```python
x,_=make_blobs(n_samples=300,centers=5,cluster_std=0.6,random_state=0)
```

```python
x_train,x_test=train_test_split(x,test_size=0.2,random_state=42)
```

```python
kmeans=KMeans(n_clusters=5,random_state=42)
kmeans.fit(x_train)
```

```python
y_train_pred=kmeans.predict(x_train)
y_test_pred=kmeans.predict(x_test)
```

```python
train_silhouette=silhouette_score(x_train,y_train_pred)
test_silhouette=silhouette_score(x_test,y_test_pred)
```

```python
print(f"Training Silhouette score: {train_silhouette}")
print(f"Test Silhouette score: {test_silhouette}")
```

```python
plt.scatter(x_train[:,0],x_train[:,-1],c=y_train_pred,s=50,cmap='viridis',label='training data')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='red',label='centroids')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```
