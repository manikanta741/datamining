```python
import matplotlib.pyplot as plt
exp_vals=[1000,500,400,300]
exp_labels=["food","rent","electricity","edu"]
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,shadow="True")
plt.show()
```
